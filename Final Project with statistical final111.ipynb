## II. Problem Modeling

### 2.1 Definition

The problem we are addressing in this study involves finding an optimal delivery route that connects a subset of cities and returns to its starting point while minimizing the total cost of the route.

In our case, we want to study the Vehicle Routing Problem (VRP), which involves finding the optimal set of routes for a fleet of vehicles to serve a set of customers, subject to various constraints such as capacity, time windows, and service times. In our case, we can consider each city as a customer that needs to be served by a single vehicle, and the objective is to minimize the total cost of the route, which can include factors such as fuel consumption, distance, and time.


We have modeled the problem for a subgraph corresponding to a single truck using the Traveling Salesman Problem (TSP) with additional constraints. Essentially, we treat the route of each truck as a separate TSP, where the truck needs to visit a set of customers and return to its starting point. In addition to the standard TSP requirements, we also include constraints. This approach allows us to optimize the route for each truck individually while ensuring that all constraints are satisfied. 


To tackle this problem, we will model it using a graph. In graph theory, a graph is a mathematical structure consisting of nodes (also called vertices) and edges that connect them. In our case, the cities will be represented as nodes in the graph, and the roads connecting the cities will be represented as edges. Each edge will have a weight associated with it, which corresponds to the cost it takes to travel between two cities.

<ins>**Proof that this decision problem is NP-Complete**<ins/>

To prove that the given decision problem is NP-complete, we can follow the two steps below:

***1) Show that the problem is in NP:***

To show that the problem is in NP, we need to demonstrate that, given a certificate (a solution), we can verify the solution in polynomial time.

In this case, a certificate would be a set of routes for each of the $K$ trucks such that:

 - Each city is visited exactly once.
 - The routes start and end at $v_0$.
 - The sum of the distances traveled by all the trucks is less than or equal to $D$.
    
We can verify this certificate in polynomial time by:

 - Checking that the length of the routes is equal to the number of cities.
 - Ensuring that there are no repeated cities in the routes (except for $v_0$, which is the starting and ending point).
 - Confirming that the sum of the distances of the routes is less than or equal to $D$.

Therefore, the problem is in NP.

***2) Show that the problem is NP-hard:***

To show that the problem is NP-hard, we can reduce a known NP-complete problem to our problem in polynomial time. The problem of the Traveling Salesman Problem (TSP) is a well-known NP-complete problem that we can use for this reduction.

The TSP can be defined as follows:

 - Data: A set of cities, the distances between each pair of cities, and a positive integer $K$.
 - Question: Is there a tour of length less than or equal to $K$ that visits each city exactly once and returns to the starting city?
    
We can reduce the TSP to our problem in polynomial time by:

Setting the number of trucks in our problem to 1.
Setting the capacity of the truck to be large enough to accommodate all the cities.
Setting the total distance $D$ in our problem to be $K$.
With these modifications, our problem becomes equivalent to the TSP. Therefore, the problem is NP-hard.

Since we have shown that the problem is in NP and NP-hard, we can conclude that the given decision problem is NP-complete.

### 3.1 Adjacency matrix

An adjacency matrix is a way of representing a graph, which is a set of objects (called "vertices" or "nodes") connected by links (called "edges" or "arcs"). In an adjacency matrix, the graph is represented by a square matrix where the rows and columns are both labeled by the vertices of the graph.

The entry in the i-th row and j-th column of the matrix (which is often denoted as $A[i,j]$) is typically either 1 or 0, depending on whether there is an edge from the i-th vertex to the j-th vertex. If the graph is undirected (meaning the edges have no orientation, i.e., they are bidirectional), then the matrix is symmetric, meaning $A[i,j] = A[j,i]$.

Here's a simple example of a weighted graph representing real-world distances:

Let's say we have an undirected graph with three vertices: $A$, $B$, and $C$. These vertices could represent three cities, and the edges between them could represent the roads or railways connecting them.

- There is an edge between $A$ and $B$ with a weight of 2. This weight corresponds to the distance between city $A$ and city $B$, which is 2 kilometers (or miles, depending on your chosen unit).
- There is an edge between $A$ and $C$ with a weight of 3. This weight corresponds to the distance between city $A$ and city $C$, which is 3 kilometers.
- There is an edge between $B$ and $C$ with a weight of 1. This weight corresponds to the distance between city $B$ and city $C$, which is 1 kilometer.

The adjacency matrix for this weighted graph would be a 3x3 matrix, like this:

$$\begin{pmatrix}
0 & 2 & 3\\
2 & 0 & 1\\
3 & 1 & 0
\end{pmatrix}$$

In this matrix:

- $A[A,B]$ = 2 because there is an edge between $A$ and $B$ with a weight of 2, which corresponds to the real-world distance of 2 kilometers.
- $A[B,C]$ = 1 because there is an edge between $B$ and $C$ with a weight of 1, which corresponds to the real-world distance of 1 kilometer.
- The diagonal entries (like $A[A,A]$) are all 0 because there are no edges from a vertex to itself in this graph. This could also mean that the distance from a city to itself is 0.

```python
import random
import numpy as np
#from functools import lru_cache
from collections import deque
import matplotlib.pylab as plt

# Function to generate random coordinates and calculate distances
def generate_distances(n_customers):
    n_points = n_customers + 1  # Plus one for the depot
    coords = np.random.randint(0, 5000, size=(n_points, 2))  # Generate random coordinates
    coords[0] = [0, 0]  # Set depot coordinates to [0, 0]
    distances = np.linalg.norm(coords[:, np.newaxis] - coords[np.newaxis, :], axis=2).astype(int)  # Calculate distances
    return distances

n_cities = 100
distances = generate_distances(n_cities)
print(distances)
```

generate_distances(n_customers): Generates random coordinates for a given number of customers and the depot. It also calculates the distances between all the points and returns the distances matrix and the coordinates data.

```python
num_truck = 4

# Function to randomly assign truck capacities
def random_capacity(n_cities, num_truck):
    if num_truck > n_cities:  # Check if input parameters are valid
        raise ValueError("Invalid input parameters")

    capacities = [1] * num_truck  # Initialize truck capacities to 1

    remaining_capacity = n_cities - num_truck  # Calculate remaining capacity to be assigned
    if remaining_capacity > 0:  # If there's remaining capacity to be assigned
        for _ in range(remaining_capacity):
            truck_index = random.randint(0, num_truck - 1)  # Select a truck randomly

            # Increase the capacity of the selected truck
            capacities[truck_index] += 1

    # Create a dictionary of truck capacities
    capacity_dict = {i + 1: capacity for i, capacity in enumerate(capacities)}

    return capacity_dict
```

random_capacities(n_cities, num_trucks): Generates random capacities for a given number of trucks, ensuring that the total capacity is equal to the number of cities (customers).

```python
# Function to generate random subgraphs for each truck
def random_subgraph(n_cities, num_truck):
    truck_capacity = random_capacity(n_cities, num_truck)  # Get the truck capacities
    cities = list(range(n_cities))  # Create a list of all cities (excluding the depot)
    print('Truck capacities:', truck_capacity)  # Print the truck capacities
    print('Total cities:', cities)  # Print the list of cities
    sub_graphs = {}  # Create an empty dictionary to store the subgraphs

    for num_truck, capacity in truck_capacity.items():  # Iterate over the truck capacities
        cities_truck = [0]  # Create a list to store the cities assigned to the current truck (including the depot)
        while len(cities_truck) < capacity:  # While the number of cities assigned to the truck is less than its capacity
            city_selection = random.randint(1, len(cities) - 1)  # Select a city from the list of cities randomly
            cities_truck.append(cities[city_selection])  # Add the selected city to the list of cities assigned to the truck
            cities.pop(city_selection)  # Remove the selected city from the list of cities
        sub_graphs[num_truck] = cities_truck  # Add the list of cities assigned to the truck to the dictionary of subgraphs
    return sub_graphs  # Return the dictionary of subgraphs

sub_graphs = random_subgraph(n_cities, num_truck)  # Generate the random subgraphs for each truck
```

random_subgraphs(n_cities, num_trucks): Generates random subgraphs for a given number of trucks and their capacities. Each subgraph represents a truck's route, starting and ending at the depot (city 0).

```python
# Function to generate a random solution for each subgraph
def random_solution(sub_graphs):
    solutions = {}
    for num_graph, graph_cities in sub_graphs.items():
        cities = graph_cities[1:]  # Exclude the depot
        random.shuffle(cities)
        cities = [graph_cities[0]] + cities + [graph_cities[0]]  # Add the depot at the beginning and end
        solutions[num_graph] = tuple(cities)
    return solutions

rand_sol = random_solution(sub_graphs)
print(rand_sol)
```

```python
tabu_list_size = 5
iter_max = 30
morebest = 0

def tabu_search(initial_solution, tabu_list_size, iter_max):
    nb_iter = 0
    # Create the tabu list as a deque with a maximum length
    tabu_list = deque((), maxlen=tabu_list_size)
   
    best_solution = initial_solution
    current_solution = initial_solution
    global_best_solution = initial_solution
    current = deque(())
    best_current = deque(())


    # Calculate the initial distance and set it as the global best distance so far
    global_best_distance = np.sum(distances[initial_solution[:-1], initial_solution[1:]])

    while nb_iter < iter_max:
        nb_iter += 1  # Increment the iteration counter 
        # Set the value of the best neighbor as infinity
        best_distance = float('inf')
  
        for neighbor in neighbour(current_solution):
            # Calculate the distance of the neighbor
            neighbor_distance = np.sum(distances[neighbor[:-1], neighbor[1:]])
            if neighbor_distance < best_distance and neighbor not in tabu_list:  
                best_solution = neighbor
                best_distance = neighbor_distance
                
        if best_distance < global_best_distance:
            global_best_solution = best_solution
            global_best_distance = best_distance
            nb_iter = 0
       

       
        best_current.append(morebest) 
        current.append(best_distance)
        
        current_solution = best_solution
        # Add the current solution to the tabu list
        tabu_list.append(current_solution)

    return global_best_solution, global_best_distance, best_current, current
```
